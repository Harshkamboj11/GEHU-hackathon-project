```python
import pandas as pd
import matplotlib.pyplot as plt  # Corrected the import alias from ptl to plt

# Specify the path to your CSV file
csv_file_path = 'detected_objects.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path) 

# Display the DataFrame
print(df)
print(df.describe())

# Plotting a bar graph for Object vs Count
plt.figure(figsize=(10, 6))  # Set the figure size
plt.bar(df['Object'], df['Count'], color='skyblue')  # Create a bar graph

# Adding title and labels
plt.title('Detected Objects vs Count')
plt.xlabel('Objects')
plt.ylabel('Count')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()  # Adjust layout to make room for labels
plt.show()
```
